```python
#time delta
df = pd.DataFrame(
        {
        'item': ['item1', 'item2', 'item3', 'item4', 'item5', 'item6'],
        'purchase_dt': pd.date_range('2021-01-01', periods=6, freq='D', tz='UTC')
        }
)
df
```

```python
df['expiration_dt'] = df['purchase_dt'] + pd.Timedelta(days=30) #구매일자에서 30일을 더하자
df
```
